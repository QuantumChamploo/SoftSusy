```python

%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image, display
import ipywidgets as widgets
```

```python
def f(mMess, TanBeta):
    print('mMess is' + str(mMess))
    string_hld = "plot"
    mHld = str(int((mMess - 4)/2))
    if(TanBeta < 7.5):
        tanb_hld = 1
    elif(TanBeta<15):
        tanb_hld = 2
    else:
        tanb_hld = 3
    string_hld += mHld
    string_hld += "."
    string_hld += str(tanb_hld)
    string_hld += ".png"
    try:
        display(Image(filename='../pics/'+string_hld))
    except:
        print("error")
```

```python
interactive_plot = interactive(f, mMess=widgets.IntSlider(min=6,max=14,step=2,value=8), TanBeta=(5, 20))
```

```python
output = interactive_plot.children[-1]
output.layout.height = '550px'
interactive_plot
```
